## 9.Mini ML Tasks


```python
# Example dataset
X = np.array([
    [50, 30],
    [60, 35],
    [70, 40],
    [80, 45]
])   # Features: [weight, age]

y = np.array([0, 0, 1, 1])   # Labels
```

```python
# feature Normalization
mean = X.mean(axis=0)
std = X.std(axis=0)

X_norm = (X - mean) / std
print(X_norm)
```

```python
# Computing Euclidean Distance
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
d = euclidean_distance(X_norm[0], X_norm[1])
print(d)
```

```python
# Implementing Prediction Step
w = np.array([0.5, -0.3])
b = 0.1
y_pred = X_norm@w +b
print(y_pred)
```

```python
# Loss Function (MSE)
mse=np.mean((y-y_pred)**2)
mse
```

```python
#Sigmoid Function
def sigmoid(z):
    return 1/1+(np.exp(-z))
y_probability=sigmoid(y_pred)
print(y_probability)
```
